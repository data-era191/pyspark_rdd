```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder.appName("find_google_word_associated _with_ip_adresses").getOrCreate()
```

```python
rdd0=spark.read.text("data.txt").rdd
rdd0.collect()
```

```python
rdd1=rdd0.filter(lambda line:"www.google.com" in line[0])
rdd1.collect()
```

```python
rdd2=rdd1.map(lambda line:line[0].split()[0])
rdd2.collect()
```

```python
unique=[rdd2.collect()[0]]
count=0
for ip1 in range(1,len(rdd2.collect())):
    for ipf in unique :
        if (rdd2.collect()[ip1] != ipf):
            count=count+1
        print(count,rdd2.collect()[ip1],unique)
    if (count == len(unique)):
        unique.append(rdd2.collect()[ip1])
        count=0
print(unique)
```
