```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder.appName("maximum_value_PM10").getOrCreate()
```

```python
rdd0=spark.read.text("data.txt").rdd
rdd0.collect()
```

```python
rdd1=rdd0.map(lambda line:float(line.value.split(",")[2]))
rdd1.collect()
```

```python
max=rdd1.collect()[0]
for i in range(1,len(rdd1.collect())):
    if(rdd1.collect()[i]>max):
        max=rdd1.collect()[i]
max 
```

```python
rdd2 = spark.sparkContext.parallelize([max])
rdd2
```

```python
rdd2.saveAsTextFile("output")
```
