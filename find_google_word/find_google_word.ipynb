```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
    .appName("find_google_word") \
    .getOrCreate()
```

```python
rdd0=spark.read.text("data.txt").rdd
rdd0.collect()
```

```python
rdd1 = rdd0.filter(lambda line:"google" in line[0])
rdd1.collect()
```

```python
rdd1.saveAsTextFile("output")
```
