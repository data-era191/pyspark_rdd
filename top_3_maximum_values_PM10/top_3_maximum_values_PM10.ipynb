```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder.appName("top_3_maximum_values_PM10").getOrCreate()
```

```python
rdd0=spark.read.text("data.txt").rdd
rdd0.collect()
```

```python
rdd2=rdd0.map(lambda line:float(line[0].split(",")[2]))
rdd2.collect()
```

```python
rdd3=rdd2.sortBy(lambda x: x, ascending=False)
rdd3.collect()
```

```python
rdd4=spark.sparkContext.parallelize(rdd3.top(3))
rdd4.collect()
```

```python
rdd4.saveAsTextFile("output")
```
